```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
from statsmodels.stats.proportion import proportions_ztest
from matplotlib.pyplot import figure
figure(figsize=(16, 12))
```

```python
ii_out = -1
n_draws_vec = range(1,100)
n_simulacoes = len(n_draws_vec)
vec_pvalue = [None]*n_simulacoes
vec_media = [None]*n_simulacoes
vec_x = [None]*n_simulacoes
n_simul = 2
for n_draws in n_draws_vec:    
    ii_out+=1
    n_draws = (n_draws*10)**2
    pval_acum = 0
    soma_acum = 0
    for ii in range(0,n_simul):
        controle= np.random.randint(0,2,n_draws)
        tratamento= np.random.randint(0,2,n_draws)
    
        soma_controle = sum(controle)
        soma_tratamento = sum(tratamento)

        count = np.array([soma_controle, soma_tratamento])
        nobs = np.array([n_draws, n_draws])

        stat, pval = proportions_ztest(count, nobs)
        pval_acum += pval
        soma_acum += soma_tratamento - soma_controle
    pval_final = pval_acum / n_simul 
    soma_final = soma_acum / n_simul 
    
    vec_x[ii_out] = n_draws
    vec_media[ii_out] = (soma_final)/n_draws
    vec_pvalue[ii_out] = pval_final
    
plt.plot(vec_x, vec_media, 'r') # plotting t, a separately 
plt.plot(vec_x, vec_pvalue, 'b') # plotting t, b separately 
plt.show()
```
