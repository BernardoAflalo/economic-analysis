```python
import random
import numpy as np
import pandas as pd
import math
from scipy import stats
import matplotlib.pyplot as plt
from statsmodels.stats.proportion import proportions_ztest
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt

%matplotlib inline
figure(figsize=(16, 12))
```

```python
controle = pd.Series()
tratamento = pd.Series()
pontos_por_iteracao = 50
qtdade_iteracoes = 200

convers_tratamento = 0.04
convers_controle = 0.04

vec_x = [None] * qtdade_iteracoes
vec_media = [None] * qtdade_iteracoes
vec_pvalue = [None] * qtdade_iteracoes

for ii_out in range(0,qtdade_iteracoes):    

    tratamento = tratamento.append(pd.Series(random.choices([1] * int(convers_tratamento*100) + [0] * int((1-convers_tratamento)*100), k=pontos_por_iteracao)))
    controle = controle.append(pd.Series(random.choices([1] * int(convers_controle*100) + [0] * int((1-convers_controle)*100), k=pontos_por_iteracao)))
    
    t2, pval = stats.ttest_ind(controle,tratamento)
    diff_soma = tratamento.sum() - controle.sum()

    
    vec_x[ii_out] = len(tratamento)
    vec_media[ii_out] = (diff_soma)/len(tratamento)
    vec_pvalue[ii_out] = pval
    
plt.plot(vec_x, vec_pvalue, 'b') # plotting t, b separately 
plt.show()
```
